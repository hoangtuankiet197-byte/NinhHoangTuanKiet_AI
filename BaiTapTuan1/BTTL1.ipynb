```python
#bai 14
import random
import math

# Nhập kích thước
m = int(input("Nhập số dòng m: "))
n = int(input("Nhập số cột n: "))

# a. Tạo mảng ngẫu nhiên
a = [[random.randint(1, 10) for _ in range(n)] for _ in range(m)]

print("\nMa trận:")
for row in a:
    print(row)

# b. Xuất dòng k
k = int(input("\nNhập dòng k: "))
print("Dòng k:", a[k])

# c. Xuất cột k
k = int(input("Nhập cột k: "))
print("Cột k:")
for i in range(m):
    print(a[i][k])

# d. Dòng có tổng lớn nhất
max_sum = -math.inf
row_index = 0
for i in range(m):
    s = sum(a[i])
    if s > max_sum:
        max_sum = s
        row_index = i
print("Dòng có tổng lớn nhất:", row_index)

# e. Cột có tích nhỏ nhất
min_prod = math.inf
col_index = 0
for j in range(n):
    prod = 1
    for i in range(m):
        prod *= a[i][j]
    if prod < min_prod:
        min_prod = prod
        col_index = j
print("Cột có tích nhỏ nhất:", col_index)

# f. Phần tử dòng chẵn, cột lẻ
print("Phần tử dòng chẵn, cột lẻ:")
for i in range(0, m, 2):
    for j in range(1, n, 2):
        print(a[i][j], end=" ")
print()

# g. TBC phần tử chẵn thuộc dòng lẻ
tong = dem = 0
for i in range(1, m, 2):
    for j in range(n):
        if a[i][j] % 2 == 0:
            tong += a[i][j]
            dem += 1
print("TBC chẵn dòng lẻ:", tong/dem if dem else 0)

# h. TBC phần tử biên
tong = dem = 0
for i in range(m):
    for j in range(n):
        if i == 0 or i == m-1 or j == 0 or j == n-1:
            tong += a[i][j]
            dem += 1
print("TBC biên:", tong/dem)

# i. TBC phần tử không biên
tong = dem = 0
for i in range(1, m-1):
    for j in range(1, n-1):
        tong += a[i][j]
        dem += 1
print("TBC không biên:", tong/dem if dem else 0)
```

```python
#bai 15
class SinhVien:
    def __init__(self, ma, ten, namsinh, dtb):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.dtb = dtb

# Nhập danh sách sinh viên
n = int(input("\nNhập số sinh viên: "))
ds = []

for i in range(n):
    print(f"\nSinh viên {i+1}:")
    ma = input("Mã SV: ")
    ten = input("Tên: ")
    namsinh = int(input("Năm sinh: "))
    dtb = float(input("Điểm TB: "))
    ds.append(SinhVien(ma, ten, namsinh, dtb))

# c. Đủ điều kiện lên lớp (>=5)
print("\nSố SV đủ điều kiện:", sum(1 for sv in ds if sv.dtb >= 5))

# d. SV đủ 20 tuổi
nam_ht = 2025
print("SV đủ 20 tuổi:")
for sv in ds:
    if nam_ht - sv.namsinh == 20:
        print(sv.ten)

# e. SV hệ DH (ký tự 2,3 là DH)
print("Số SV hệ DH:", sum(1 for sv in ds if sv.ma[2:4] == "DH"))

# f. Tên Lan
print("Số SV tên Lan:", sum(1 for sv in ds if "Lan" in sv.ten))

# g. Họ Phan
print("Số SV họ Phan:", sum(1 for sv in ds if sv.ten.startswith("Phan")))
```
