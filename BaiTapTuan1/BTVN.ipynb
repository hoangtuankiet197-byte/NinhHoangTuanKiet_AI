```python
#bai 1:
def tron_mang(a, b):
    c = []
    i = 0
    while i < len(a) and i < len(b):
        c.append(min(a[i], b[i]))
        i += 1
    if i < len(a):
        c.extend(a[i:])
    if i < len(b):
        c.extend(b[i:])
    return c

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print(tron_mang(a, b))
```

```python
#bai 2:
def tong_tam_giac_tren_cheo_phu(a, n):
    s = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                s += a[i][j]
    return s


def doi_am(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
    return a


def thay_chan(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x
    return a


def toan_chan(a):
    for row in a:
        for val in row:
            if val % 2 != 0:
                return False
    return True


def doi_xung(a, n):
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True


def cheo_chinh_tang(a, n):
    for i in range(n - 1):
        if a[i][i] > a[i + 1][i + 1]:
            return False
    return True


def xuat_tam_giac_duoi_cheo_phu(a, n):
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()


def cheo_phu_giam(a, n):
    for i in range(n - 1):
        if a[i][n - 1 - i] < a[i + 1][n - 2 - i]:
            return False
    return True

a = [
    [1, 2, 3],
    [4, -5, 6],
    [7, 8, 9]
]
n = 3

print("Tổng tam giác trên chéo phụ:", tong_tam_giac_tren_cheo_phu(a, n))

print("Đổi âm:")
print(doi_am(a))

print("Thay số chẵn bằng 0:")
print(thay_chan(a, 0))

print("Toàn chẵn:", toan_chan(a))

print("Đối xứng:", doi_xung(a, n))

print("Chéo chính tăng:", cheo_chinh_tang(a, n))

print("Tam giác dưới chéo phụ:")
xuat_tam_giac_duoi_cheo_phu(a, n)

print("Chéo phụ giảm:", cheo_phu_giam(a, n))
```
